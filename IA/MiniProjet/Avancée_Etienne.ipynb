Importation des données 


```
from sklearn.datasets import load_diabetes
data=load_diabetes()
```

```
data.DESCR
```

```
data.target
```

```
dataBase=data.data
print(dataBase)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(dataBase,data.target,test_size=0.4, train_size=0.6);
X_test, X_validation,y_test,y_validation = train_test_split(X_test,y_test, test_size=0.5, train_size=0.5); 

print("X_train de taille:",len(X_train),X_train)
print("X_test de taille:",len(X_test),X_test)
print("y_train de taille:",len(y_train),y_train)
print("y_test de taille:",len(y_test),y_test)
print("X_validation de taille:",len(X_validation),X_validation)
print("y_validation de taille:",len(y_validation),y_validation)
```

## SVM Linéaire 

```
# Imports
import numpy as np
from sklearn import datasets,linear_model
from sklearn import svm

# Linear Support Vector Classification.
clf_1=svm.LinearSVR(C=10)
clf_1.fit(X_train,y_train)

# Epsilon-Support Vector Regression.
clf_2= svm.SVR(kernel='linear', C=10)
clf_2.fit(X_train,y_train)

# Linear Support Vector Machine  linear
clf_3 = svm.SVC(kernel = 'linear', random_state = 0)
clf_3.fit(X_train,y_train)

# Radial Support Vector Machine rbf 
clf_4 = svm.SVC(kernel = 'rbf', random_state = 0)
clf_4.fit(X_train,y_train)




```

Prédiction 

```
print("predict CLF1")
pred_clf1 = clf_1.predict(X_test)
print(pred_clf1)

print("predict CLF2")
pred_clf2 = clf_2.predict(X_test)
print(pred_clf2)

print("predict CLF3")
pred_clf3 = clf_3.predict(X_test)
print(pred_clf3)

print("predict CLF4")
pred_clf4 = clf_4.predict(X_test)
print(pred_clf4)
```

Evaluation courbe ROC


```
from sklearn.metrics.cluster import normalized_mutual_info_score


clf1_NMI= normalized_mutual_info_score(y_test,pred_clf1)
print("NMI clf1 :",clf1_NMI)

clf2_NMI= normalized_mutual_info_score(y_test,pred_clf2)
print("NMI clf2 :",clf2_NMI)

clf4_NMI= normalized_mutual_info_score(y_test,pred_clf4)
print("NMI clf2 :",clf2_NMI)

clf4_NMI= normalized_mutual_info_score(y_test,pred_clf4)
print("NMI clf4 :",clf4_NMI)
```
